```python
import pandas as pd
import numpy as np
import os
import pickle

from typing import Any
```

```python
arquivos = []
for arquivo in os.listdir("data"):
    caminho_arquivo = os.path.join("data", arquivo)

    with open(caminho_arquivo, "rb") as f:
        raw_data:list[dict[str,Any]] = pickle.load(f)
        arquivos.append(raw_data)



# arquivo é uma lista com 500 dicionarios e cada dicionario é uma linha do df final
```

```python
numero, disperso, concentrado = [], [], []

for arquivo in arquivos:
    for coluna in arquivo[0].keys():
        valor = arquivo[0][coluna]
        tipo = type(valor).__name__
        media = np.mean(valor)
        std = np.std(valor)
        texto = f"{coluna}: {tipo}    {media}     {std}"
        print(texto)

        if tipo == "float":
            numero.append(coluna)
        elif std > media:
            disperso.append(coluna)
        else:
            concentrado.append(coluna)

# Alguns std sao maiores que a media, nesses casos usarei a mediana. Nos demais usarei a media
```

```python
# Suposicao

print(numero)
print(disperso)
print(concentrado)
```

```python
cols = ['rajd', 'decjd', 'bary_period', 'topo_period', 'dm', 'snr', 'width', 'subints', 'subbands', 'profile', 'dm_index', 'period_index', 'accn', 'hits', 'rank', 'fftsnr', 'dm_values', 'dm_curve_snr_values', 'accn_values', 'accn_curve_snr_values']
pre_df = []

numero = ['rajd', 'decjd', 'bary_period', 'topo_period', 'dm', 'snr', 'width', 'accn', 'hits', 'rank', 'fftsnr']
disperso = ['subints', 'subbands', 'profile', 'period_index', 'accn_values']
concentrado = ['dm_index', 'dm_values', 'dm_curve_snr_values', 'accn_curve_snr_values']


for arquivo in arquivos:
    for dici in arquivo:
        linha = []
        for key, value in dici.items():
            if key in numero:
                if isinstance(value, float):
                    linha.append(1)
                else:
                    linha.append(0)

            elif key in disperso:
                if np.std(value) >= np.mean(value):
                    linha.append(1)
                else:
                    linha.append(0)

            elif key in concentrado:
                if np.std(value) < np.mean(value):
                    linha.append(1)
                else:
                    linha.append(0)

        pre_df.append(linha)


df = pd.DataFrame(pre_df, columns=cols)

df.mean()

# colunas que nao se enquadram em disperso ou nao: period_index (68%)
```
